```python
# remove unrated restaurants
df_rated = df[df['Aggregate rating'] > 0]

# keep only cuisines with at least 10 restaurants
popular_cuisines = df_rated.groupby('Cuisines').filter(lambda x: len(x) >= 10)

# calculate average rating
popular_cuisine_rating = popular_cuisines.groupby('Cuisines')['Aggregate rating'].mean()

# sort results
popular_cuisine_rating.sort_values(ascending=False)
```

Modern Indian cuisine has the highest average rating, indicating strong customer satisfaction.

Traditional cuisines like Indian, Seafood, and Thai also receive high ratings from customers.

Some mixed cuisine combinations involving fast food and multiple cuisines tend to have slightly lower ratings.

Restaurants specializing in fewer cuisines often perform better in ratings compared to those offering many different cuisines.

```python
# Step 1: remove unrated
df_rated = df[df['Aggregate rating'] > 0]

# Step 2: filter popular cuisines
popular_cuisines = df_rated.groupby('Cuisines').filter(lambda x: len(x) >= 10)

# Step 3: calculate average rating
popular_cuisine_rating = popular_cuisines.groupby('Cuisines')['Aggregate rating'].mean()

# Step 4: SORT and STORE
top_cuisines = popular_cuisine_rating.sort_values(ascending=False).head(10)

# Step 5: plot
top_cuisines.plot(kind='bar')

plt.title("Top 10 Cuisines by Average Rating")
plt.xlabel("Cuisine")
plt.ylabel("Average Rating")

plt.show()
```

### Restaurant Chain Analysis

```python
chain_counts=df['Restaurant Name'].value_counts()
chain_counts
chain=chain_counts[chain_counts>1]
chain
```

#### Top Chain Restaurants

```python
Top_chain=chain.head(10)
Top_chain
```

```python
# Average rating of chain Restaurants
chain_rating=df.groupby('Restaurant Name')['Aggregate rating'].mean()
chain_rating
```

#### Analyze Ratings of Chains

```python
chain_rating=chain_rating[chain_rating.index.isin(chain.index)]  #“Is this restaurant name in the list of chains?” 
#eg Domino's Pizza → True
#KFC → True
# Random Cafe → False
chain_rating.sort_values(ascending=False).head(10)
```

#### Analyze Popularity (Votes)

```python
chain_votes=df.groupby('Restaurant Name')['Votes'].mean()
chain_votes=chain_votes[chain_votes.index.isin(chain.index)]
chain_votes.sort_values(ascending=False).head(10).round(2)
```

```python
chain_analysis = df[df['Restaurant Name'].isin(chain.index)]
#chain.index → list of restaurant chains (names appearing more than once)
#.isin() → checks if restaurant name is in that list
# Single restaurants are removed
chain_summary = chain_analysis.groupby('Restaurant Name').agg({
    'Aggregate rating': 'mean',
    'Votes': 'mean',
    'Restaurant ID': 'count'  # number of outlets or branches
}).rename(columns={'Restaurant ID': 'Number of Outlets'})

chain_summary.sort_values(by='Aggregate rating', ascending=False).head(10) # sorting based on rating
```

```python
top_chain_ratings = chain_summary.sort_values(by='Aggregate rating', ascending=False).head(5)

top_chain_ratings['Aggregate rating'].plot(kind='bar')

plt.title("Top Restaurant Chains by Rating")
plt.xlabel("Restaurant Chain")
plt.ylabel("Average Rating")

plt.xticks(rotation=45)
plt.show()
```
